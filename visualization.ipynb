### 이미지에 대한 데이터프레임 시각화

```python
import cv2
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
from func import *
```

```python
column_names = ['img_name', 'average', 'sd', 'score']

val_df = pd.read_csv('val_df.csv', names=column_names, header=0)
bright_df = pd.read_csv('bright_df.csv', names=column_names, header=0)
dark_df = pd.read_csv('dark_df.csv', names=column_names, header=0)
```

```python
tmp = bright_df.sort_values(by='average', ascending=True).reset_index(drop=True)
tmp['sd'].plot()
plt.title('밝은 이미지에서 밝기 증가에 따른 sd 변화')
```

```python
tmp = dark_df.sort_values(by='average', ascending=True).reset_index(drop=True)
tmp['sd'].plot()
plt.title('어두운 이미지에서 밝기 증가에 따른 sd 변화')
```

```python
plt.figure(figsize=(15,6))
plt.subplot(1, 2, 1)
plt.plot(val_df['average'])
txt1 = "average 평균: " + str(round(val_df.average.mean(), 3))
plt.xlabel(txt1)
plt.title('score로 정렬한 전체 이미지의 average')

plt.subplot(1, 2, 2)
plt.plot(val_df['sd'])
plt.title('score로 정렬한 전체 이미지의 sd')
txt2 = "sd 평균: " + str(round(val_df.sd.mean(), 3))
plt.xlabel(txt2)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
plt.plot(dark_df['average'])
txt1 = "average 평균: " + str(round(dark_df.average.mean(), 3))
plt.xlabel(txt1)
plt.title('score로 정렬한 어두운 이미지의 average')

plt.subplot(1, 2, 2)
plt.plot(dark_df['sd'])
txt2 = "sd 평균: " + str(round(dark_df.sd.mean(), 3))
plt.xlabel(txt2)
plt.title('score로 정렬한 어두운 이미지의 sd')
plt.show()
```
