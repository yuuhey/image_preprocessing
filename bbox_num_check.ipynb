```python
import json

# json 데이터로부터 한 이미지 내 bbox 개수 확인 함수
def print_filenames_with_multiple_bboxes(json_data):
    for image_info in json_data:
        image_name = image_info['image_id']  # 이미지 파일 이름은 'image_id'로 대체
        bboxes = image_info.get('keypoints', [])

        if len(bboxes) > 1:
            print(f"Image: {image_name} has multiple bboxes.")
            # for bbox_index, bbox in enumerate(bboxes, start=1):
            #     print(f"  Bbox {bbox_index}: {bbox}")
        elif len(bboxes) == 1:
            pass
        else:
            print(f"Image: {image_name} has no bbox.")

# JSON 파일 경로
json_file_path = 'result_keypoints_small_base.json'

# JSON 파일 읽기
with open(json_file_path, 'r') as f:
    data = json.load(f)

# bbox 여부 출력하는 함수 사용
print_filenames_with_multiple_bboxes(data)
```
